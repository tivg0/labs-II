```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.odr import RealData, Model, ODR
import sys
sys.path.append("../../..")
from baseLabs import *
```

```python
betaesperado = 300

def uV(V,div):
    return abs(V)*.03 + 0.1*div + 0.001

def uBeta(dec,Rb,Rc,udec):
    return np.sqrt((Rb/Rc*udec)**2+(dec/Rc*Rb*.05)**2+(dec*Rb/Rc**2*Rc*.05)**2)
```

```python
Vin, Vout, Rc, Rb = getData("1.txt")


mask = Vin > -3.8
Vin = Vin[mask]
Vout = Vout[mask]

uVin = uV(Vin,1)
uVout = uV(Vout,1)

adjust = fullLinAnalysis(Vin,Vout,uVin,uVout,r"V_{out} (V_{in})", r"V_{in} (V)",r"V_{out} (V)", tol=2,separate=False,s=3)

Rc = Rc[0]
Rb = Rb[0]
beta = adjust.beta[0]*Rb/Rc
betaerr = abs(beta+betaesperado)/betaesperado*100
ubeta = uBeta(adjust.beta[0],Rb,Rc,adjust.sd_beta[0])
print(f"β = {-beta:.0f} ± {ubeta:.0f} \nε_rel = {betaerr:.2f} %\n")
dec, incdec, trash = round_un(adjust.beta[0],adjust.sd_beta[0])
print(f"declive = {-dec} ± {incdec}")
```
